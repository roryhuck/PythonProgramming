```python
#getting html
url = "https://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Number of quotes on main page
quotes = soup.find_all("div", class_="quote")
len(quotes)
```

```python
# Creating lists of the quotes, authors and tags
quote_texts = []
authors = []
tags = []

for q in quotes:
    quote_texts.append(q.find("span", class_="text").text)
    authors.append(q.find("small", class_="author").text)
    
    for t in q.find_all("a", class_="tag"):
        tags.append(t.text)

print("Quotes:", quote_texts)
print("Authors:", authors)
print("Tags:", tags)
```

```python
tag_df = pd.DataFrame({"Tag": tags})
print("Duplicate tags", tag_df.duplicated().sum())
```

```python
tag_dict = {}

for tag in tag_df["Tag"].unique():

    page = requests.get(f"https://quotes.toscrape.com/tag/{tag}/")
    soup = BeautifulSoup(page.text, "html.parser")

    tag_dict[tag] = [(q.find("span", class_="text").text,
                      q.find("small", class_="author").text)
                     for q in soup.find_all("div", class_="quote")[:10]]
```
