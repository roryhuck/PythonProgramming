2 questions we are going to try to answer using this data:
1. Which franchise has the most releases?
2. Which franchise has the highest total domestic gross?


```python
# Question 1
max_releases = franchises_df.loc[franchises_df['Releases'].idxmax()]

print("Franchise with the most releases:")
print(max_releases)
```

```python
# Question 2
franchises_df['Total'] = (
    franchises_df['Total']
    .replace('[\$,]', '', regex=True)
    .astype(float)
)

max_gross_franchise = franchises_df.loc[franchises_df['Total'].idxmax()]

print("Franchise with highest total gross:")
print(max_gross_franchise)
```

```python
print(f"Most Releases: {max_releases['Franchise']} ({max_releases['Releases']} films)")
print(f"Highest Total Gross: {max_gross_franchise['Franchise']} (${max_gross_franchise['Total']:,.0f})")
```
