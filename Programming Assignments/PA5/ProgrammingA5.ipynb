```python
import random

trivia ={}

with open('AnimalTrivia.txt') as f:
    for line in f:
        line = line.strip()
        if line.startswith('#Q'):
            q = line[2:].strip()
            trivia[q] = {'a': '', 'c': []}
        elif line.startswith('^'):
            trivia[q]['a'] = line[1:].strip()
        else:
            trivia[q]['c'].append(line)

right = 0
wrong = 0

while input('Play some Triva? (yes/no): ').lower() == 'yes':
    q = random.choice(list(trivia))
    print('\n' + q)
    for choice in trivia[q]['c']:
        print(choice)

    if input('Answer: ').lower() == trivia[q]['a'].lower():
        print('Correct!')
        right += 1
    else:
        print('Incorrect! Answer was:', trivia[q]['a'])
        wrong += 1

with open('TriviaResults.txt', 'a') as f:
    f.write(f'Correct: {right}, Incorrect: {wrong}\n')

print('Done')
```
