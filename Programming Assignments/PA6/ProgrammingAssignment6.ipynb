### Analyze which beverage category has:
- The most calories per drink?
- The most caffiene per drink?
- The most sugar per drink?
- The most proten per drink?

This will probably require you to write a for loop for each beverage in the beverage categories list.

In the loop you will need to get just the drinks that match each category from the dataframe and then sum up the calories, caffiene, sugar, and protein, for those drinks. Then divide by the total number of drinks in the category.

You will also need to keep track of the most calories, caffiene, sugar, and protein as you loop through to check the values. If your sum is greater then store both the amount and the category.

```python
max_cal = 0
max_caff = 0
max_sugar = 0
max_protein = 0

cal_cat = ''
caff_cat = ''
sugar_cat = ''
protein_cat = ''

for c in cat:
    
    drinks = df[df['Beverage_category'] == c]
    
    avg_cal = drinks['Calories'].mean()
    avg_caff = drinks['Caffeine (mg)'].mean()
    avg_sugar = drinks['Sugars (g)'].mean()
    avg_protein = drinks['Protein (g) '].mean()

    if avg_cal > max_cal:
        max_cal = avg_cal
        cal_cat = c

    if avg_caff > max_caff:
        max_caff = avg_caff
        caff_cat = c

    if avg_sugar > max_sugar:
        max_sugar = avg_sugar
        sugar_cat = c

    if avg_protein > max_protein:
        max_protein = avg_protein
        protein_cat = c


print('Max calories:', cal_cat)
print('Max caffeine:', caff_cat)
print('Max sugar:', sugar_cat)
print('Max protein:', protein_cat)
```

```python
categories = df['Beverage_category'].unique().tolist()
columns = ['Calories', 'Sugars (g)', 'Protein (g) ', 'Caffeine (mg)', 'Total Fat (g)']

print('Beverage categories:', ', '.join(categories))
choice = input('Choose a beverage category: ').strip().lower()
category_map = {c.lower(): c for c in categories}
if choice not in category_map: exit()
choice = category_map[choice]
print(f'You chose: {choice}')

print('Nutrition columns:', ', '.join(columns))
column = input('Choose a nutrition column: ').strip().lower()
column_map = {c.lower(): c for c in columns}
if column not in column_map: exit()
column = column_map[column]

try:
    max_val = float(input(f'Enter maximum {column}: '))
except: exit()

filtered = df[(df['Beverage_category'] == choice) & (df[column] <= max_val)]

if len(filtered) == 0:
    print(f'No drinks under {max_val} {column} in {choice}.')
else:
    filtered = filtered.sort_values(column)
    print(filtered[['Drink', column]])
    if input('Save to my_drinks.csv? (yes/no): ').strip().lower() == 'yes':
        filtered.to_csv('my_drinks.csv', index=False)
        print('Saved')
```
