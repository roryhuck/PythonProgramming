### A useful operation: Modulo

In many places clocks report the time from 0-24 hours so that you don't have to keet track of am vs pm. For example 8am = 8:00 and 8pm = 20:00. We will use modulo to make a time converter.

```python
# Modulo
prompt = 'Enter the hours on your 24 hour clock: '
hours = int(input(prompt))
prompt = 'Enter the minutes on your 24 hours clock: '
minutes = input(prompt)

twelve_hour_time = hours % 12

if twelve_hour_time == 0:
    twelve_hour_time = 12

if hours < 12:
    ampm = 'am'
elif hours >= 12:
    ampm = 'pm'

print(f'The time is {twelve_hour_time}:{minutes}{ampm}.')
```

### While Loop

While loop? While not!

These are great for continuing a process until a condition is met.

```python
# Enter what classes you are taking:
prompt = "\nEnter another class you are taking:"
classes = []

while len(classes) < 4: 
    class_name = input(prompt)
    classes.append(class_name)
    
print(f'\Here is your list of classes: {classes}')
```

```python
# But wait!
# What if someone is only taking three classes or is actually taking 5 classes?
# What if they accidentally enter the same class twice?

prompt = "\nEnter another class you are taking:"
prompt += "\nWhen you are done entering classes type 'stop'."
classes = []
class_name = ''
while class_name != 'stop':
    class_name = input(prompt)
    if class_name != 'stop':
        if class_name not in classes:
            classes.append(class_name)
        else:
            print(f'You already entered {class_name}.')
    
    
print(f'\Here is your list of classes: {classes}')
```

### Flag, Break, and Continue

   - You can use a flag variable to run a while loop until a boolean goes from True to False
   - You can use the 'break' command to stop a loop in it's tracks
   - You can use a 'continue' command to skip some parts of a loop

```python
# FLAGS
# Here is a guessing game
# The flag got_it only changes from False to True if the person makes the right guess
# The while loop keeps going as long as got_it is False.

my_word = 'bat'
prompt = "Enter a three letter word:"
prompt += "\nType 'quit' to give up."

print('Try to guess my three letter animal:')
got_it = False #Flag
num_guesses = 0

while not got_it:
    word = input(prompt)
    if len(word) != 3:
        if word == 'quit':
            got_it = True
            print(f'    You gave up after {num_guesses} guesses')
        else:
            print("    You wasted a guess - that word was not three letters!")
            num_guesses += 1
    elif word == my_word:
        print(f"    You got it in {num_guesses+1} guesses! {word} was my three letter animal")
        got_it = True
    else:
        num_guesses += 1
        print(f"   Wrong guess! You have made {num_guesses} guesses")
```

```python
# BREAK
# Lets you stop code in its tracks!
# In this code we break the while loop once the user says quit.
colors = []
print('Enter as many colors as you can think of.')

prompt = 'Enter a color'
prompt += ' or type "quit":'

while True:
    color = input(prompt)
    if color == 'quit':
        break
    else:
        if color in colors:
            print(f'\nYou already guessed {color}')
        else:
            colors.append(color)
            print(f'\nSo far you have: {colors}.')
print('-----------------------------')
print(f'You finished with {len(colors)} colors:')
print(colors)
```

```python
# CONTINUE
# Lets the loop keep going without finishing the current run
# In this code we continue to the next prompt unless there are three people in the single_group list
prompt = "How many people per group. \nEnter a number:"
num_g = int(input(prompt))

single_group = []
all_groups = []
prompt = 'Enter a name to add to the group \n type "quit" when you are done.'

while True:
    name = input(prompt)
    if name == 'quit':
        break
    single_group.append(name)
    if len(single_group) % num_g != 0:
        continue 
    else:
        all_groups.append(single_group)
        single_group = []

i=0
while len(single_group) != 0:
    all_groups[i].append(single_group.pop())
    i += 1
    
print(all_groups)
```

### There are lots of other advanced things you can do with while loops!

They interact well with lists and dictionaries!


```python
# Say I want to move names from one list to another, but check that the names are not repeated.
# Like Programming Assignment 3.
# Here we use while to keep running until a list is empty.

current_names = ['Joanna','Ross','Stefanie']
current_ages = [44,44,75]
new_names = ['Joanna','Bella','Kristin']
new_ages = [44,18,40]

while new_names:
    name = new_names.pop()
    if name in current_names:
        continue
    else:
        age = new_ages.pop()
        if age > 21:
            current_names.append(name)
            current_ages.append(age)
        else:
            print(f"{name} is too young!")
            
print(current_names)
print(new_names)
```

```python
# Say I have a list of users, but somehow a bunch of names got repeated!
names = ['Joanna', 'Ross', 'Stefanie', 'Kristin']
# Artificially ruin the list:
import numpy as np
for i in range(20):
    num = np.random.randint(len(names))
    names.append(names[num])
    
print("My list of usernames is all messed up. There are repeated names!\n")    
print(names)

print('-------------------------')
# Now let's fix this mess
repeates = True
unique_names = []
while repeates:
    if len(names) == 0:
        repeates = False
    else:
        name = names.pop()
        while name in names:
            names.remove(name)
        unique_names.append(name)
        
print('Here is the final list with just the unique names:\n')
print(unique_names)        
```
