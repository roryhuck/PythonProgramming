### Day 23 - Object Oriented Programming (OOP) and Python Classes

Object oriented programming (OOP) is one of the best approaches to programming, especially as your programs get more complex. In this style of programming you write **Classes**. The idea is that classes should represent real world things. Then you define **Objects** that are members of the class. The objects behave similarly to everyting in their class.

Now this sounds really abstract and there is some extra vocab we might need. But once you get the hang of it, it's really useful!

#### Vocab

- Instantiation - creating an object in a class
- Instance - you work with instances of a class.
- Attributes - variables that are available to each instance.
- Method - a function defined within a class.
- Inheritence - the ability of one class to inherit properties from another.


#### This will all become more clear with an example

```python
# Create a class of cats!
class Cat:
    '''A model of a cat.'''
    # Initialize the cat using the __init()__ method
    # TWO UNDERSCORES!!!!
    def __init__(self,name,age,color):
        '''Initialize the attributes of the cat
           all cats in the class have these attributes.
           The user specifies these valies when they create a cat'''
        self.name = name
        self.age = age
        self.color = color
        
    # All the things the cat can do - methods (functions)
    def eat(self):
        print(f'{self.name} is now eating')
        
    def praise(self):
        print(f'{self.name} you are such a wonderful {self.color} cat!')
        
    def birthday(self):
        self.age += 1
        print(f'Happy Birthday {self.name} you are now {self.age} years old.')
```

#### When I run this code nothing happens we just store the idea of a cat in pythons brain.

This class definition has a few important parts:

Above you see the word init() with two underscores before and after it. The init() method with two underscores is a special thing that Pyton runs whenever we create a new instance (a new cat). Here are some important things about init()

- The self parameter is required and it must come first
- All other parameters are optional and based on the thing your are describing
- The other parameters are defined as self.parameter, this lets us pass that information to ever function as part of self - they are attributes.

After we instantiate (create) our instance (cat) the object (the cat) has attributes (name,age,color) but it also has defined methods (functions inside the class). These functions can be anything that ever member of the class would do. Our cat has three methods:

- eat
- praise
- birthday
