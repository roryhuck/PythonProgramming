### Day 17 - Web Scraping - Getting Data from the Web

Today we are going to learn a few different ways to get data from a Website into Python. Web scraping takes quite a bit of patience. Every website is different and they way they store their data is different. So for each site you will have to logically figure out how to access the data.

#### Ethical Issues

There are some things to be aware of before you start scraping data from the web. 

- Some data is private or protected. Just because you have access to a websites data doesn't mean you are allowed to scrape it. For example, when you log into Facebook or another social media site, you are granted special access to data about your connected people. It is unethical to use that access to scrape their private data!

- Some websites have rules against scraping and will cut of service to users who are clearly scraping data. How do they know? Webscrapers acess the website very differently that regular users. If they site has a policy about scraping data then you should follow it and/or content them about getting the data if you have a true academic interest in the data.

- The line between web scraping and plaigerism can be very blurry. Make sure that you are citing where your data comes from AND not just reproducing the data exactly. Always citing the source of your data and make sure you are doing something new with it.

- Ethics are different depending on if you are using the data for a personal project (eg. you just want to check scores for your favorite team daily and print the stuff you care about) vs if you are using the project for your business or website (eg. publishing information to drive clicks to your site/video/account or making money from the data you colloce). In the later case it is EXTRA important to respect the original owner of the data. Drive web traffic back to their site, check with them about using their data, etc.

The Ethical Scraper (from https://towardsdatascience.com/ethics-in-web-scraping-b96b18136f01):

I, the web scraper will live by the following principles:

- If you have a public API that provides the data I’m looking for, I’ll use it and avoid scraping all together.
- I will always provide a User Agent string that makes my intentions clear and provides a way for you to contact me with questions or concerns.
- I will request data at a reasonable rate. I will strive to never be confused for a DDoS attack.
- I will only save the data I absolutely need from your page. If all I need it OpenGraph meta-data, that’s all I’ll keep.
- I will respect any content I do keep. I’ll never pass it off as my own.
- I will look for ways to return value to you. Maybe I can drive some (real) traffic to your site or credit you in an article or post.
- I will respond in a timely fashion to your outreach and work with you towards a resolution.
- I will scrape for the purpose of creating new value from the data, not to duplicate it.

3. Here are some ideas for questions:

4. Can I come up with a new statistic for how to count the 'best" player?


```python
# Let's look at the unnammed columns
columns = ['Unnamed: 5', 'Unnamed: 22','Unnamed: 26']

# How many rows are in the data?
rows = len(ACES)
print(f'There are {rows} rows.')

# How many NaNs are in the unnamed rows
for c in columns:
    num_na = sum(ACES[c].isna())
    print(f'Column {c} has {num_na} NANs.')
```

```python
# So there is nothing in the unnamed columns!
# All of the rows are NAN.
```

```python
# Let's answer the question: Who has played the most years, games, minutes?

columns = ['Yrs','G','MP']
mosts = ['years','games','minutes played']
display_num = 5

i=0
for c in columns:
    DF_tmp = ACES.sort_values(c,ascending=False).head(display_num)
    print(f'The top {display_num} players in terms of {mosts[i]}')
    display(DF_tmp[['Player',c]])
    i+=1
```

## Using Beautiful Soup to get HTML code

#### The code below will work well for static websites - aka sites that don't use fancy things like javascript to actively load content.

Websites are built using html code. That code tells the web browser (FireFox, Chrome, etc) what to display. Websites can be very simple (just html) to much more complicated (java script +). When you load a website you can always see the source code:

- Right Click - view page source

This is what beautiful soup downloads. For static (simple) sites this code is immediately available. More complicated sites might require Python to open the webpage, let the content render, and then download the code.



### How to get data from static sites:

```python
import requests
from bs4 import BeautifulSoup
```

```python
website = "https://joannabieri.com/python.html"
```

```python
raw_code = requests.get(website)
html_doc = raw_code.text
soup = BeautifulSoup(html_doc, 'html.parser')
```

#### Lets see what is in soup

```python
soup
```

```python
soup.find_all?
```

### This is HTML code

#### WHAT A MESS!

The information in soup is ALL of the code and unless you are awesome at reading HTML, this is indecipherable. We need to be able to find specific parts of this to extract the data.

### Extracting data:

Here is the full function signature:

    soup.find_all(name=None,attrs={},recursive=True,text=None,limit=None,**kwargs,)
    
You can type soup.find_all? and run it to see all the inforrmation. Here is how we will mostly use it, but there are much more advnaced things you can do:

    soup.find_all( <type of section>, <info> )
    
The .find_all() function searches through the information in soup to match and return only sections that match the info. Here are some important types you might search for:

- 'h2' - this is a heading
- 'div' - this divides a block of information
- 'span' - this divides inline information
- 'a' - this specifies a hyperlink
- 'li' - this is a list item

- class_= - many things have the class label (notice the underscore!)
- string= - you can also search by strings.


### Using Developer tools:

To figure out what data to extract I suggest you use developer tools on the website to find what you need. Navigate to the website:

https://joannabieri.com/python.html

The class lecture video will help walk you through this process.


```python
# Here we search for all the links
# Links are in the section of the HTML marked with <a
result = soup.find_all('a')
print(result)
```

```python
# This is still a mess but the results can be looped through
# We can look just at the text inside each block ignoring the extra html stuff:
for r in result:
    print(r.text)
```

```python
# We can look just at the just the link information inside each block ignoring the rest
for r in result:
    print(r.get('href'))
```

### Lets write code that gets all of the links and then does stuff with those links

```python
# Here we search for all the links
result = soup.find_all('a')
# Save all the data to lists
# One for the html links
# One that containts the text
'''
The line below is a for loop using list comprehension.
It does the same as:
   
   class_links = []
   for r in result:
       class_links.append(r.get('href'))

List comprehension is often faster than looping because of internal Python optimization.
'''
class_links = [r.get('href') for r in result]
link_text = [r.text for r in result]
```

### You can search for sections inside sections. 

Maybe you only want things that are inside both a div and a span. Let's look through the website and find information about what we did in each day of class.

Notice on the developer tools. This information is inside a \<div\> and inside a \<span\>

```python
# Here we search for words about each class day
# Get the results for searching inside both div and span
result = soup.find_all(['li','div','span'])
# Put all the words into a list using list comprehension.
words = [r.text for r in result]
print(len(words))
for w in words:
    if "CLASS" in w:
        print('----------------')
        print(w.replace('\n',''))
```
