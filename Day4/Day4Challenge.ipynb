#### Loop examples 

```python
my_list = [3,8,7,2,9]
for i in range (len(my_list)):
    print(my_list[i])
```

```python
my_colors = ['red','orange','yellow','green','blue','purple']

for i in (my_colors):
    print(i)
```

```python
my_colors = ['red','orange','yellow','green','blue','purple']
my_endings = ['st','nd','rd','th','th','th']
for i in range (len(my_colors)):
    print(f' My {i+1}{my_endings[i]} color is {my_colors[i]}.')
```

```python
my_colors = ['red','orange','yellow','green','blue','purple']
string = ''
for i in my_colors:
    string += i+' '
print(string)
```
