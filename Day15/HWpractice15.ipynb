#### Arbitrary Number of Arguments.

Sometimes you just don't know how many arguments your user will input:

- Some people have just a first and last name, some people have middle names to, some people have family names that pass through generations and could have many names.

- You might not know how many items are on a persons grocery list, this changes week to week. One week you might get 3 things, the next week 30.

The special character * allows you to send in an arbitrary number of items

```python
def my_breakfast(*foods):
    print(f'For Breakfast I had: {foods}')
    
# REMEMBER - when you run a cell that just contains a function 
#            it will look like nothing happened!
#            You are storing that function in Pythons brain
```

```python
my_breakfast('smoothie','toast','coffee')
```

```python
# I could redo my function and make it print more nicely with a FOR LOOP
def my_breakfast(*foods):
    print('For Breakfast I had:')
    for food in foods:
        print(f"   {food}")
```

#### 8. Write a function using an arbitrary argument:

Write a function that takes in items that a person wants to put on their Sandwich. The function should contain one parameter can accept an arbitrary number of items. The function should print out a summary of the order.

- eg1 If I run sandwich_order('ham','cheese','mayo') it should return: "Your order has been recived, you will have ham, cheese and mayo on your sandwich".

- eg2 If I run sandwich_order('cheese') it should return: "Your order has been recived, you will have cheese on your sandwich".

```python
# Your code here
def sandwich_order(*items):
    if not items:
        print('You did not order anything.')
        return
    if len(items) == 1:
        order_str = items[0]
    else:
        order_str = ', '.join(items[:-1]) + ' and ' + items[-1]
    print(f'Your order has been received, you will have {order_str} on your sandwich')
sandwich_order('turkey','cheese','mayo','lettuce','tomato')
sandwich_order('cheese')
```

#### Combining Parameter Types.

You often want to combine different parameter types in a function. You should be careful of order here:

- FIRST you should define positional arguments
- SECOND you should define keyword arguments
- THIRD you should define arbitrary arguments
- FOURTH you should define all default arguments

```python
def my_meal(meal,*foods):
    print(f'For {meal} I had:')
    for food in foods:
        print(f"   {food}")
```

```python
my_meal('Breakfast','smoothie','toast','coffee')
```

```python
my_meal('Lunch','soup','salad','soda')
```

```python
def dramatic_excuse(name, problem="my homework", *reasons, intensity=1):
    message = "Dear Professor,\n\n"
    if reasons:
        reason_text = " and ".join(reasons)
        message += f"I could not complete {problem} because {reason_text}.\n"
    else:
        message += f"I could not complete {problem}.\n"
    if intensity > 1:
        message += "It was " + "VERY " * intensity + "unfortunate" + "!" * intensity + "\n"
    message += f"\nSincerely,\n{name}"
    return message
```
