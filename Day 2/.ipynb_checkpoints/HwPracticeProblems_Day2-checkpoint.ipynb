### Data Types - Strings

```python
mystring = 'This is my awesome STRING!'
print(mystring)
```

```python
# There are lots of things that you can do with strings:

# Make all the first letters upper case
print(mystring.title())

# Make every letter upper case
print(mystring.upper())

# Make every letter lower case
print(mystring.lower())

# Remove first word
print(mystring.removeprefix("This"))

# Remove last few characters
print(mystring.removesuffix('ING!'))

# Swap out letter
mystringTwo = mystring.replace('i','******************')
print(mystringTwo)
```

```python
# Try using the tab key
mystring.lower()
```

```python
# You can even merge two or more strings together:
mystring1 = "Wow this is a really"
mystring2 = "really really really really really really long string!"

combined_string = f"{mystring1} {mystring2}"
print(combined_string)



# And it doesn't matter what you use as variable names:
batman = "Wow this is a really"
superman = "really really really really really really long string!"

catwoman = f"{batman} {superman}"
print(catwoman)
```

```python
# All of these ideas can be combined to create new strings
# Let's play madlibs!
first_name = "Kristin"
last_name = "Bieri"
favorite_food = "mashed potatoes"
favorite_class = "Python"
action = "skip"
exclamation = "Oh Honest to Pete!!"


#
#
#
#

my_paragraph = f"Once when {first_name} {last_name} was in her most favorite class, {favorite_class}, she happend to {action} while eating a big plate of {favorite_food}! The whole class screamed {exclamation}! After that no one was allowed to bring {favorite_food} to {favorite_class} class"
print(my_paragraph)
```

```python
# If you have really long strings you can break them up
my_paragraph = f"Once when {first_name} {last_name} was in her most favorite class, {favorite_class},"\
               f" she happend to {action} while eating a big plate of {favorite_food}!"\
               f" The whole class screamed {exclamation}!"\
               f" After that no one was allowed to bring {favorite_food} to {favorite_class} class"
print(my_paragraph)
```

### Numbers

```python
# Python can do quite a bit with basic numbers
a = 1345
b = 20
print("add:", a+b)
print("multiply:", a*b)
print("subtract:", a-b)
print("divide:", a/b)
```

```python
# Raising things to a power is a little weird
print("power:", a**b)
```

For more complicated matematics we need to import a package called numpy. This package contains all sorts of more complicated math equations.

```python
# This is how you import a package
import numpy as np

# If python started up by loading every package it would take FOREVER to load!
# This lets you choose what you really need to make your code run.
```

```python
# Now that we have numpy loaded we can do harder math
print(np.sin(a))
```

```python
print(np.sqrt(a))
```
