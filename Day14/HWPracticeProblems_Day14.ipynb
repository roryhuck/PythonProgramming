Remeber in programming assignment 5 when we were looking at Starbucks Drinks and I asked you to find:

#### Which beverage category has:
- The most calories per drink?
- The most caffiene per drink?
- The most sugar per drink?
- The most proten per drink?

You probably wrote something like this:

```python
# Import the data in the file
import pandas as pd
file_name = 'https://joannabieri.com/python/starbucks_drinkMenu_expanded.csv'
DF = pd.read_csv(file_name)

bev_cats = list(DF['Beverage_category'].unique())

most_cals = 0
most_caf = 0
most_sug = 0
most_pro = 0

for c in bev_cats:
    # c is a beverage category
    mask_cats = DF['Beverage_category']==c
    DF_cat = DF[mask_cats]
    total_drinks = len(DF_cat)
    
    total_cals = sum(DF_cat['Calories'])
    cal_per_drink = total_cals/total_drinks
    if cal_per_drink > most_cals:
        most_cals = cal_per_drink
        most_cal_category=c
        
    total_caf = sum(DF_cat['Caffeine (mg)'])
    caf_per_drink = total_caf/total_drinks
    if caf_per_drink > most_caf:
        most_caf = caf_per_drink
        most_caf_category=c
        
    total_sug = sum(DF_cat['Sugars (g)'])
    sug_per_drink = total_sug/total_drinks
    if sug_per_drink > most_sug:
        most_sug = sug_per_drink
        most_sug_category=c
        
    total_pro = sum(DF_cat['Protein (g) '])
    pro_per_drink = total_pro/total_drinks
    if pro_per_drink > most_pro:
        most_pro = pro_per_drink
        most_pro_category=c
        
print(f'{most_cal_category} have the most calories per drink at {most_cals}')
print(f'{most_caf_category} have the most mg caffeine per drink at {most_caf}')
print(f'{most_sug_category} have the most sugar per drink at {most_sug}')
print(f'{most_pro_category} have the most protein per drink at {most_pro}')
```

```python
# I try to import the data just once at the top
import pandas as pd
file_name = 'https://joannabieri.com/python/starbucks_drinkMenu_expanded.csv'
DF = pd.read_csv(file_name)
```

```python
def get_most(col_name,DF):
    '''Function to calculate the drink category that has the most
       of the category in col_name'''
    bev_cats = list(DF['Beverage_category'].unique())
    most = 0
    for c in bev_cats:
        mask_cats = DF['Beverage_category']==c
        DF_cat = DF[mask_cats]
        total_drinks = len(DF_cat)

        total = sum(DF_cat[col_name])
        per_drink = total/total_drinks
        if per_drink > most:
            most = per_drink
            most_category=c
            
    print(f'{most_category} has the most {col_name} at {most} per drink.')
```

```python
get_most('Calories',DF)
```

```python
# Your code here
def top_movies_by_genre(genre, n=5):
    if genre not in DF['genre'].unique():
        print("That genre does not exist in the dataset.")
        return
    
    genre_movies = DF[DF['genre'] == genre]
    genre_movies = genre_movies.sort_values(by='inflation_adjusted_gross', ascending=False)
    top_movies = genre_movies.head(n)
    
    print(f"We found {len(top_movies)} movies.")
    print(top_movies[['movie_title','inflation_adjusted_gross']])
```

## Returning Values from a Function

```python
def make_email(first,last):
    '''Creates an email address for first last'''
    my_email = f'{first.lower()}_{last.lower()}@redlands.edu'
    return my_email
```

### Practice 6

Define a function that returns the solution to the pathagorean formula:

$$ a^2 + b^2 = c^2 $$

It should have three optional parameters (a,b,c) and then depending on which two are given returns the solution for the third.

For example:

- import numpy as np
- a=2, b=3
- the code: if c is not specified then solve ans = np.sqrt(a^2+b^2)
- return the answer.

It should let the user send in any combo of two constants AND give an error if the user sends in less then two or more than two.



```python
# Your code here
import numpy as np
def pythagorean(a=None, b=None, c=None):
    values = [a, b, c]
    
    if values.count(None) != 1:
        print("Error: provide exactly two values.")
        return
    
    if c is None:
        return np.sqrt(a**2 + b**2)
    
    if a is None:
        return np.sqrt(c**2 - b**2)
    
    if b is None:
        return np.sqrt(c**2 - a**2)
print(pythagorean(a=2,b=3))
```
