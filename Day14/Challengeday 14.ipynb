```python
def c_to_f(celsius):
    return (9/5) * celsius + 32
# test
print(c_to_f(0))
print(c_to_f(20))
print(c_to_f(100)) 
```

```python
def ticket_price(age):
    if age <= 12:
        return 8
    elif age <= 64:
        return 12
    else:
        return 10

print(ticket_price(10))
print(ticket_price(30))
print(ticket_price(70))
```

```python
families = {
    "Huck": [19, 50, 52 ],
    "Bieri": [36, 40, 10, 12],
    "Smith": [50, 45, 12]
}

for family_name, ages in families.items():
    total_cost = 0
    for age in ages:
        total_cost += ticket_price(age)
    print(f"Total cost for the {family_name} family is ${total_cost}")
```
