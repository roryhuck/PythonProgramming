### Creating a Pandas DataFrame from scratch

This is a lot like a dictionary,

```python
# The basic structures in pandas are dataframes
# Think of them like a dictionary
my_dict = {'Column1':['r1','r2','r3'],
           'Column2': ['r1','r2','r3']}

example_df = pd.DataFrame(my_dict,index=['Row1','Row2','Row3'])
display(example_df)
```

### Accessing the data in the DataFrame

```python
# By Index Location (row)
display(example_df.iloc[0])
```

```python
# By Name (row)
display(example_df.loc['Row1'])
```

```python
# By Name (column)
display(example_df['Column1'])
```

### Adding data to the DataFrame

```python
# Adding columns
example_df['Column3'] = ['r1','r2','r3']
display(example_df)
```

```python
# Adding rows at a specific location
example_df.loc['Row4'] = ['r4','r4','r4']
example_df
```

### Changing the names of rows and columns:

```python
# You can update the name of a row
example_df = example_df.rename({'Row4':'The Best Row'})
example_df
```

```python
# You can update the name of all rows
new_names = {'Row1':'A',
             'Row2':'B',
             'Row3':'C',
             'The Best Row':'D'}
example_df = example_df.rename(new_names)
example_df
```

```python
# You can change the name of the columns
example_df.columns = ['A','B','C']
example_df
```

### Finding NaNs

NaN is a special entry basically meaning that nothing was entered or that something incorrect or indecipherable was entered. It can be thouhg of as infinity which is Not A Number = NaN. Now it is used more generally. 

```python
# Pandas can find these

# Create a mask
nan_mask = Disney_DF['mpaa_rating'].isna()

# Now use the mask on the DataFrame
Disney_DF[nan_mask]
```

```python
# How many NaNs are there in the data:
num_nans = nan_mask.sum()
print(f"There are {num_nans} NaNs in our data!")
```

### What do you do with these NaNs?

This is actually a huge decision!!! Are NaNs bad or do they hold some meaning? Here are our options.

- Leave them in the data and interpret what they mean.
- Drop them all
- Other or a combination (drop some)

```python
# Here is how we could drop all the nans
Disney_DF_noNAN = Disney_DF.dropna()
Disney_DF_noNAN
```

```python
# Check if there are any Nans left:
Disney_DF_noNAN['mpaa_rating'].isna().sum()
```

### Challenge Problem

Make a bar graph the shows the total inflation adjusted gross for Disney movies by year. Here is an outline of what you might do.

- First loop through  each value in the release date column to create a column with just the year (we did this above so this step is done)
- Then figure out how many unique years there are.
- Loop through this list of years and add up the inflation adjusted gross for each year
- After you have the data create the bar graph.


```python
# Your code here
years = list(Disney_DF['release_year'].unique())

year_totals = []

for y in years:
    total = 0
    for i in range(len(Disney_DF)):
        if Disney_DF['release_year'].iloc[i] == y:
            total += Disney_DF['inflation_adjusted_gross'].iloc[i]
    year_totals.append(total)
```

```python
plt.figure()
plt.bar(years, year_totals)

plt.xlabel("Year")
plt.ylabel("Total Inflation Adjusted Gross")
plt.title("Total Inflation Adjusted Gross by Year")

plt.xticks(rotation=90)  
plt.show()
```
