### Day 24 - More OOP and Class Inheritence

The nice thing about classes is that you don't always have to start from scratch, you can use classes that you have already written and inherit from them.

For example, what if I had a class called Cat and it had all sorts of things about a Cat:

- Attributes: Name, Size, Color
- Methods: eat(), sleep(), pounce(), run()

But then I wanted to define a class for a Kitten.... well a kitten is a special type of cat, so I could inherit properties from Cat when I am writing my Kitten class.


#### This will all become more clear with an example

#### Here is a class of cats from last time

```python
# Create a class of cats!
class Cat:
    '''A model of a cat.'''
    # Initialize the cat using the __init()__ method
    # TWO UNDERSCORES!!!!
    def __init__(self,name,age,color):
        '''Initialize the attributes of the cat
           all cats in the class have these attributes.
           The user specifies these valies when they create a cat'''
        self.name = name
        self.age = age
        self.color = color
        
    # All the things the cat can do - methods (functions)
    def eat(self):
        print(f'{self.name} is now eating')
        
    def praise(self):
        print(f'{self.name} you are such a wonderful {self.color} cat!')
        
    def birthday(self):
        self.age += 1
        print(f'Happy Birthday {self.name} you are now {self.age} years old.')
```

#### Let's make a cat!

```python
my_cat = Cat('Bongo',14,'black')
```

```python
print(my_cat.name)
print(my_cat.age)
print(my_cat.color)
```

#### Let's make the cat do things!

```python
my_cat.eat()
my_cat.praise()
my_cat.birthday()
```

```python
my_cat.age
```

#### Now lets make a kitten which we recognize as a special kind of cat

We will send the class "Cat" into the kitten class and then initialize it using the init() from the Cat class....

```python
class Kitten(Cat):
    '''A model of a kitten which is a type of cat'''
    def __init__(self,name,age,color):
        '''Initialize the attributes of the parent class'''
        super().__init__(name,age,color) # This creates a cat - but it is also a kitten
    
    # The methods here will only apply to kittens!
    def tiny_meow(self):
        print(f'{self.name} just made a TINY MEOW!')
```

```python
baby_cat = Kitten('Bella',0.5,'black')
```

```python
print(baby_cat.name)
print(baby_cat.age)
print(baby_cat.color)
baby_cat.eat()
baby_cat.tiny_meow()
```

Notice that we cannot make Bongo do a tiny_meow() since that method is only part of the Kitten class.

```python
my_cat.tiny_meow()
```

The kitten can do all the things the cat can do, but with special attributes or methods added to it.

#### There is no limit to the number of ways you can customize your classes!

In the example below we will give the kitten some extra attributes and overwrite some of the cat functions.

```python
class Kitten(Cat):
    '''A model of a kitten which is a type of cat'''
    def __init__(self,name,age,color,mom):
        '''Initialize the attributes of the parent class'''
        super().__init__(name,age,color) # This creates a cat - but it is also a kitten
        self.mom = mom
    
    # The methods here will only apply to kittens!
    def tiny_meow(self):
        print(f'{self.name} just made a TINY MEOW!')
        
    # This overwrites the eat method for any instance of kitten.    
    def eat(self):
        print(f'The kitten {self.name} is drinking milk from mom {self.mom}')
```

```python
baby_cat = Kitten('Bella',0.5,'black','Muffin')
baby_cat.eat()
```

```python
my_cat.eat()
```

### Notes about Classes

As you write more and more code and create bigger classes sometimes your code can get really long! This makes organizing your code really important. Here are some suggestions:

1. Classes should all be named using CamelCase. Capitalize the first letter of each word with no spaces.
2. You should put just one space betweem methods in a Class
3. You should put two spaces between different Classes in a block of code.

#### Modules

Just like with functions we can store classes in modules and then import them. This is really nice if you have LOTS of classes and functions defined. It makes your code easier to read. 

1. Always load the standard python packages (numpy, pandas, etc) before you own.

#### If you notice that a class is getting really long or that the way your are calling different methods is confusing - break it up

This is a more advanced or challenging idea but... You can use an instance as an atribute to separate long class definitions. This is called composition - breaking a larger class into smaller classes that work together. 

Here is an examle of two classes working together:

The Employee class has information about the person and their pay. It can print the total yearly pay for the person.

The Salary class contains all the salary information with methods to get total pay and increase the salary.

Writing the classes this way is nice because 

     employee.salary.increase()
  
makes more sense than

     employee.increase()
     
You can also separate out actions that are specific to salary vs things that are specific to the employee.

```python
class Salary:
    def __init__(self,hourly_pay,hours):
        self.pay=hourly_pay
        self.hours = hours

    def get_total(self):
        weeks = 4.5
        months = 12
        return (self.pay*self.hours*weeks*months)

    def increase(self,percent):
        self.pay *= percent


class Employee:
    def __init__(self,name,hours,pay):
        self.name = name
        self.hours = hours
        self.pay=pay
        self.salary=Salary(self.pay,self.hours)

    def annual_salary(self):
        total_pay = self.salary.get_total()
        message = f'Total Yearly Pay for {self.name}: ${total_pay}'
        return message
    
    def update_pay(self):
        self.pay = self.salary.pay


first_employee=Employee('Joanna',40,15)
print(first_employee.annual_salary())
```

```python
first_employee.salary.increase(1.2)
first_employee.update_pay()

print(f'New hourly pay: {first_employee.pay}')
print(first_employee.annual_salary())
```

### Practice 2 -  Challenge

Below is code that simulates a fight between two players.

1. Can you figure out how the code works?
2. Can you add some actions and attributes:
    - the attribute magic with the method heal() - that would allow players to heal dammage
    - the attribute stealth with the method steal() - that would allow players to steal attack points
3. Can you let a user decide on whether to attack, heal, or steal durring each play while letting the opponent choose randomly?

What other ways could you change this code to make a fun game?

```python
import random
import time

FIGHTER = {"name":"fighter","health": 5, "attack": 4, "dodge": 1}
THIEF = {"name":"thief","health": 4, "attack": 3, "dodge": 5}
MAGE = {"name":"mage","health": 2, "attack": 6, "dodge": 3}

class Character:

    def __init__(self, char_type, health,attack,dodge):
        self.char_type = char_type
        self.health = health
        self.attack = attack
        self.dodge = dodge

    def attack(self):
        return self.attack

    def take_damage(self, damage):
        if self.dodge_success():
            return "Missed!"
        self.health -= damage
        return f"{self.char_type.title()} took {damage} damage."

    def dodge_success(self):
        # Every point in dodge is 10% chance to dodge
        dodge_chance = self.dodge * 10
        dodge_roll = random.randint(1, 100)
        if dodge_roll <= dodge_chance:
            return True
        return False

    def is_dead(self):
        return self.health <= 0


def character_fight(type_1, type_2):
    character_1 = Character(type_1['name'],type_1['health'],type_1["attack"],type_1['dodge'])
    character_2 = Character(type_2['name'],type_2['health'],type_2["attack"],type_2['dodge'])
    
    coin_toss = random.randint(0, 1)
    if coin_toss == 0:
        first, second = [character_1, character_2]
    else:
        first, second = [character_2, character_1]

    while True:
        if attack_character(first, second):
            print(f'{first.char_type.title()} has won!')
            return 
        if attack_character(second, first):
            print(f"{second.char_type.title()} has won!")
            return 


def attack_character(first, second):
    print(f'{first.char_type.title()} has attacked with {first.attack} attack points.')
    time.sleep(.5)
    damage = first.attack
    print(f'{second.char_type.title()} has {second.health} health.')
    time.sleep(.5)
    print(second.take_damage(damage))
    time.sleep(.5)
    if second.is_dead():
        print(f'{second.char_type.title()} is dead!')
        return True
    return False


# Main Code:
NAME = input('Enter your name: ')
WHO = int(input('Who do you want to fight? \n 1. FIGHTER, 2. THIEF, 3. MAGE\n Enter a number: '))
char_1 = {"name":NAME,"health": 10, "attack": 2, "dodge": 2}
if WHO == 1:
    char_2 = FIGHTER 
elif WHO == 2:
    char_2 = THIEF
else:
    char_2 = MAGE
    
message = f"You are fighting the {char_2['name']} who has {char_2['health']} health, {char_2['attack']} attack, and {char_2['dodge']} dodge."
print(message)
message = f"You have {char_1['health']} health, {char_1['attack']} attack, and {char_1['dodge']} dodge."
print(message)
print('------------------------------')
input('Press ENTER when you are ready to begin')

character_fight(char_1, char_2)
```
