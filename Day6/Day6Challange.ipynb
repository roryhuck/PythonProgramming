The factorial (symbol: !) means multiplying all numbers from the chosen number down to 1. For example, a factorial of 5! is 5 × 4 × 3 × 2 × 1 = 120. Make sure to check for negative numbers and non integers where we don't have a factorial and if num=0 know that 0! = 1

```python
#1
numbers = [12, 20, 75, 150, 180, 145, 525, 50]

for n in numbers:
    if n > 500:
        print(n, "is too big.")
        continue

    if n > 150:
        continue

    if n % 5 == 0:
        print(n, "divides evenly by 5")
```

```python
#3
num = 6

factorial = 1

for i in range(1, num + 1):
    factorial *= i 

print(f"The factorial of {num} is {factorial}")
```
