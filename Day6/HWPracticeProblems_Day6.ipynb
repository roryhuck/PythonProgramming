#### Checking multiple things

Often things are not so simple to just use an if-else statement. For example, maybe you want to print different things based on a wide range of ages or maybe there are different results based on different income categories. This is where elif will really help.

We are going to stay with our conditional statements for an extra day to make sure we are feeling good about these ideas. Only one new idea today.


```python
# Admissions Prices
# Disney Land - how much will I pay?
upper_age_baby = 3
upper_age_child = 9
prices = [0,50,104]

my_age = 44

if my_age < upper_age_baby:
    print('You get in for FREE!')
elif my_age >= upper_age_baby and my_age < upper_age_child:
    print(f'Your ticket will cost ${prices[1]}')
elif my_age >= upper_age_child:
    print(f'Your ticket will cost ${prices[2]}')
```

### Nesting and combining If statements

This will help you check multiple things.

```python
my_friends = ['Amaia','Diana','Margo','Kristin','Patrick','Lauren','Stefanie']
their_ages = [43,40,42,40,7,40,75]
their_locations = ["Idaho",'Idaho','Massachusets','Australia','Australia','Oregon','Idaho']

# Lets say I am having a party and only want to invite my friends who will be in the area
# and friends who are in the right age range

party_loc = 'Idaho'
lower_age = 21
upper_age = 50

# Do some checks
if len(my_friends)==len(their_ages) and len(their_ages)==len(their_locations):
    for i in range(len(my_friends)):
        if their_locations[i] == party_loc:
            if their_ages[i] >= lower_age and their_ages[i] <= upper_age:
                print(f"Hi {my_friends[i]},\n I am having a party in {party_loc}. I hope you can make it!")
        else:
            if their_ages[i] >= lower_age and their_ages[i] <= upper_age:
                print(f"Hi {my_friends[i]},\n I am having a party in {party_loc}. Do you want to join virtually?")    
else:
    print(f"Check the lenghts of your list they don't match up!\n "\
          f"{len(my_friends)}\n {len(their_ages)}\n {len(their_locations)}")
```
