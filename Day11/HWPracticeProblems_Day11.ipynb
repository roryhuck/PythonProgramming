```python
# Reading data from a file - standard way from a file in the directory
with open('my_birthday.txt', mode='r') as bday:
    for day in bday:
        print(day)  
        
# Or you can read the whole thing from a given directory path
# from pathlib import Path
# path = Path('my_birthday.txt')
# day = path.read_text()
# print(day)
```

```python
# Writing multiple lines
people = ['Mom','Dad','Kristin','Bongo']
birthdays = ['12-10','3-2','11-6','11-25']

with open('family_birthdays.txt', mode='w') as bday:
    for i in range(len(people)):
        info = f'{people[i]}:{birthdays[i]}\n'
        bday.write(info)
```

```python
sample_string = 'Joanna:2-7-1979\n'
sample_string.strip('\n').split(':')
```

```python
# Reading multiple lines and saving the data
people_new = []
birthdays_new = []

with open('family_birthdays.txt', mode='r') as bday:
    for item in bday:
        # print(item)
        # print(item.strip('\n').split(':'))
        info = item.strip('\n').split(':')
        people_new.append(info[0])
        birthdays_new.append(info[1])
print(people_new)
print(birthdays_new)
```

```python
# Appending to the end of a file
more_people = ['Ross','Bella']
more_birthdays = ['9-25','5-20']

# mode = 'a' lets you append to the bottom of the file!
with open('family_birthdays.txt', mode='a') as bday:
    for i in range(len(more_people)):
        info = f'{more_people[i]}:{more_birthdays[i]}\n'
        bday.write(info)
        
# Code coppied from above to read the file again        
people_new = []
birthdays_new = []
with open('family_birthdays.txt', mode='r') as bday:
    for item in bday:
        # print(item)
        # print(item.strip('\n').split(':'))
        info = item.strip('\n').split(':')
        people_new.append(info[0])
        birthdays_new.append(info[1])
print(people_new)
print(birthdays_new)
```

```python
# Is my birthday in the first million digits of pi?
# one-million.txt
# 1 Million Digits of Pi
# collected by Eve Andersson
# http://www.eveandersson.com/pi/digits/1000000

# Read in the birthday and remove the dashes
with open('my_birthday.txt', mode='r') as bday:
    for day in bday:
        myday = day.replace('-','')
        

# Read in the digits of pi from the file
pi_string = ''
with open('one-million.txt', mode='r') as digits:
    for d in digits:
        pi_string += d
    
# Check to see if your birthday is in there
if myday in pi_string:
    place = pi_string.find(myday)
    print(f'Found your birthday! {myday} is at location {place} in pi.')
else:
    print('Birthday not found!')
```

### Image Files

Digital images are just files that contain numbers as data. That data is then interpreted by the computer to become the pictures that we are use to seeing. But you can interact with image data just like we have been interacting with lists.

```python
import matplotlib.pyplot as plt

# Read the image from the file
my_map = plt.imread('map.png')
# Choose the size of the plot
plt.figure(figsize=(16,8))
# Plot the Image
plt.imshow(my_map)
# Show the image nicely on the screen
plt.show()
```

```python
# What is the data here?
print(my_map.shape)
# This is FOUR arrays (or multi columned lists) of data each 1275X1749 pixels rectangle.
# The four layers are Red, Green, Blue, and Transparancy
# my_map
```

Computers read in image data as colors and transparencies. Then they layer that data on top and interpret it into a image. Once you have a basic image, you can always layer more information over the top!

If you are interested it GIS or spatial studies. This is the basis for many of the maps that are curated with the ESRI software. Get a detailed map, along with data about the locations on that map. Then layer the information. Obviously they then build in lots of functions and operations!

```python
# Adding data to an image:
# We need data for the latitude and longitude of the edges of the plot
BBox = ((-117.17166, -117.15589,      
         34.05830, 34.06782))
# We need latitude and longitude data for each of our data points
Duke = (-117.1664,34.0617)

# Choose the image size
plt.figure(figsize=(16,8))
# Telll python the longitude and latitude of each corner of the map
plt.xlim(BBox[0],BBox[1])
plt.ylim(BBox[2],BBox[3])
# Create a scatter plot of points we are interested in
plt.scatter(Duke[0], Duke[1], alpha= 0.9, c='r', s=200)
# Add the plot under the points
plt.imshow(my_map, extent = BBox, aspect= 'equal')
plt.show()
```

### More fun with Images

There are lots of libraries for playing with images in python! We are going to learn about skimage here. We will import some specific tools and then use them.

Inspired by: https://towardsdatascience.com/introduction-to-image-processing-with-python-representation-of-images-for-beginners-b95725b523ca

```python
# Number and Plotting tools
import numpy as np
import matplotlib.pyplot as plt

# Image tools
from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

```python
# Here is a very simple image (checkerboard)
checkers = np.array([[255, 0, 255], 
                    [0, 255, 0],
                    [255, 0, 255]])
imshow(checkers, cmap = 'gray')
```

```python
# Here is a very simple image (checkerboard)
# But now python will think of it in colors!
colors = np.array([[[255, 0, 0], 
                         [0, 255, 0],
                         [0, 0, 255]]])
color_names = ['red','green','blue']

imshow(colors)

for i,c in enumerate(color_names):
    print(f'The matrix {colors[0][i]} gives the color {c}')
    
```

```python
# The same thing happens for real images
study_cat = imread('studycat.jpg')
imshow(study_cat)
print(f'The shape of this data is {study_cat.shape}')
```

```python
# Slice the image - just like slices of a list we can slice these arrays!
sliced_cat = study_cat[:,150:250]
imshow(sliced_cat)
```

### Practice Problem 5

Write code that will read in the Jeopardy questions data set using pandas

JEOPARDY_CSV.csv

Then do the following:

1. Save the questions to a list.
2. Save the answers to a list.
3. Write code that lets a user practice Jepordy questions
    - First it should ask if they want to play
    - If they say yes then it selects a random question and answer pair
    - Prints out the question
    - Lets the user type the answer
    - Checks the answer
    - If they get it right - congratulate them
    - If they get it wrong write (append) the question and answer to a file called practice.txt.
    - Ask them if they want to play again and keep adding to the list of practice questions.



```python
# Your Code Here
df = pd.read_csv("JEOPARDY_CSV.csv")
df
```

```python
df.columns
```

```python
questions = list(df["Question"])
answers = list(df["Answer"])
```

```python
import random # import random to choose a random question
play = input("Do you want to play Jeopardy? (yes/no): ") #user input to see if they want to play

while play.lower() == "yes": #forces all imputs to be lowercase 

    i = random.randint(0, len(questions) - 1) # defining i to be a random question

    question = questions[i] # links random question to answer
    answer = answers[i]

    print("\nYour question is:") # prints the random question
    print(question)

    user_answer = input("\nYour answer: ") # allows the user to imput the answer

    if user_answer.strip().lower() == str(answer).strip().lower(): #checks answer and takes away any weird characters/case sensitivity 
        print("Correct!") # correct answer
    else:
        print(f"Incorrect, The correct answer was: {answer}")#incorrect answer

        with open("practice.txt", "a") as file: # add missed question to practice file
            file.write(f"Question: {question}\n")
            file.write(f"Answer: {answer}\n")# adds line inbetween 

    play = input("\nWant to play again? (yes/no): ") #ask if they want to play again

print("\nThanks for playing!") #end of game 
```
