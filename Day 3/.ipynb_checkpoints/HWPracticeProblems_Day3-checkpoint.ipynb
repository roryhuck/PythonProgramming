### Day 3 - Lists

#### Defining a List and accessing the Data

```python
# A list is a collection of items in some order
# Literally think of a list on a piece of paper.

best_cat_names = ['Bongo','Tater Tot','Sofia','Bella','Muffin'] 
# Square brackets tell Python this is a list
# Separate items with commas

print(best_cat_names)
```

```python
# Now that this data is stored in the list we can access it:

print(f"Give me the first name in the list: {best_cat_names[0]}")
# Notice how we start counting the items at zero!
# Other programming languages start counting at one.

print(f"Give the the third name in the list: {best_cat_names[2]}")
```

```python
# You can also use negative indexing (the location of the intem in the list is it's index)

# Two ways to get the last item in the list:
print(f"Give me the last name in the list: {best_cat_names[4]}")
print(f"Give me the last name in the list: {best_cat_names[-1]}")
```

#### Adding new items to a list

```python
# I will start by defining an empty list
# Like a blank piece of paper

my_schedule = []

# Now add my first class

my_schedule.append("Differential Equations 1:15pm")
print(my_schedule)
```

```python
# Next add my last class
my_schedule.append("Python 6:00pm")
print(my_schedule)
```

```python
# Oh WAIT! I forgot my middle class
my_schedule.insert(1,'Senior Seminar 2:40pm')
print(my_schedule)
```

#### Deleting Items in a list

```python
# Lets remove the two duplicate classes

del my_schedule[1] # remove the second item in the list
del my_schedule[-1] # remive the last item in the list
print(my_schedule)
```

This looks much better, but notice that we lost the data completely.

```python
shopping_list = ['bread','milk','coffee','apples','cat food']
print(shopping_list)
```

#### Popping items off a list

```python
# Let's write code that removes things from the list, but remembers what we removed
# Why remember? Maybe I go to different stores and I want to keep track of the things I have already purchaced.
```

```python
# Create an empty list to keep track of purchased items
purchased_items = []

# My first stop is the pet store
purchased_items.append(shopping_list.pop())
print(f"Still on my list: {shopping_list}\n Already purchased: {purchased_items}")
```

```python
# My next top is the coffee shop
purchased_items.append(shopping_list.pop(2))
print(f"Still on my list: {shopping_list}\n Already purchased: {purchased_items}")
```

```python
# And then I realized that I already have bread!
shopping_list.remove("bread")
print(f"Still on my list: {shopping_list}\n Already purchased: {purchased_items}")
```

#### Organizing a list

```python
books_to_read = ['The Broken Earth Trilogy, N K Jemisin',
                 '1984, George Orwell',
                 'Dune, Frank Herbert',
                 'The Wisard of EarthSea, Ursula LeGuin',
                 'All Systems Red, Martha Wells',
                 'Three Body Problem, Cixin Liu']
print(books_to_read)
```

```python
# Maybe I want these listed in alphabetical order
books_to_read.sort()
print(books_to_read)
```

```python
# Maybe I want these listed in reverse alphabetical order
books_to_read.sort(reverse=True) #Notice how the word True is highlighted this is a boolean
print(books_to_read)
```

```python
books_to_read[0]
```

```python
# Maybe they were in the correct order to begin with!
# But I just want them to be printed alphabetically, leaving the list untouched.

books_to_read = ['The Broken Earth Trilogy, N K Jemisin',
                 '1984, George Orwell',
                 'Dune, Frank Herbert',
                 'The Wisard of EarthSea, Ursula LeGuin',
                 'All Systems Red, Martha Wells',
                 'Three Body Problem, Cixin Liu']
print(f"Alphabetical order: {sorted(books_to_read)}")
print("-------------------------------")
print("The original list is not changed I can still print it:")
print(f"The original order: {books_to_read}")
```
