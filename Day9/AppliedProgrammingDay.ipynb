Loop through all the names in the list to find the shortest and longest name. Print those two names

### Problem 9 - Challenge

Write a "dating app" for choosing a Harry Potter Character. Your code shoud:

- Randomly choose a character and print their name, house, skills, hair color, and eye color.
- Ask the user if they want to choose this character as their favorite.
    - If the user says yes - prints a congratulations message and ends
    - If the user says no - asks if they want to see another match.
        - If they say yes - choose another character and try again
        - If they say no - print a thanks for tryin message and ends.

```python
# Data you can use
chr_names = list(df['Name'])
chr_houses = list(df['House'])
chr_skills = list(df['Skills'])
chr_eyes = list(df['Eye colour'])
chr_hair = list(df['Hair colour'])
```

```python
import random

while True:
    play = input('Do you want to find a match? (yes/no): ')
    if play != 'yes':
        print("Thanks for trying! See you next time.")
        break

    i = np.random.randint(len(chr_names))
    print("\n--- Your Character---")
    print("Name:", chr_names[i])
    print("House:", chr_houses[i])
    print("Skills:", chr_skills[i])
    print("Hair:", chr_hair[i])
    print("Eyes:", chr_eyes[i])

    fav = input('\nDo you want this to be your favorite? (yes/no): ')
    if fav == 'yes':
        print(f"\nCongrats! {chr_names[i]} is now your favorite!")
        break
    else:
        another = input("Do you want to see another match? (yes/no): ")
        if another != 'yes':
            print("Thanks, see you next time.")
            break
```

```python
for house in houses:
    males_in_house = sum(g == "Male" for h, g in zip(chr_houses, chr_gender) if h == house)
    females_in_house = sum(g == "Female" for h, g in zip(chr_houses, chr_gender) if h == house)
    
    print(f"\nHouse: {house}")
    print("  Male:", males_in_house)
    print("  Female:", females_in_house)
```
